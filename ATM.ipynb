```python
#ATM Simulate - Python Activity
"""Rules:
• Print a main menu to ask the user different options for displaying the bank account balance, for depositing
money, to withdraw money, and to exit from the ATM.
• Additional (optional): Ask the user a PIN to access to the ATM menu once the PIN is correct otherwise ask the
user to try again.
• If ‘1’ is entered, display the current balance and return to main menu.
• If ‘2’ is entered, print a sub-menu with different withdrawal amounts (give also a choice for “other amount’ to let
the user enter an amount) :
• Check that the requested withdrawal is allowed, print a message to show that the money has been
withdrawn, calculate the new balance and return to main menu.
• If ‘other_amount’ is selected, then prompt the user for an integer value. Check this number is a multiple
of 10 and that the withdrawal is permitted, print a message to show that the money has been
withdrawn, calculate the new balance and return to main menu.
• If ‘exit’ is selected print a goodbye message and exit (break).
• If another value is entered, print an error message and print the menu again.
• Re-use the ATM menu options across the program to ask the user if he/she would like to perform another operation on
the bank account. """

account_balance = 1000.00
correct_pin = "1234"

def check_balance():
    print("Your current account balance is $", account_balance)

def deposit():
    global account_balance
    amount = float(input("Enter the amount to deposit: $"))
    account_balance += amount
    print("Deposit $", amount, "into your account.\nYour new balance is $", account_balance)

def withdraw(withdraw_amount):
    global account_balance
    if withdraw_amount > account_balance:
        print("Insufficient funds.\nYour balance is $", account_balance)
    else:
        account_balance -= withdraw_amount
        print("Withdrawn $", withdraw_amount, "from your account.\nYour new balance is $", account_balance)

def verify_pin():
    pin_attempt = input("Enter your PIN: ")
    if pin_attempt == correct_pin:
        return True
    else:
        print("Incorrect PIN. Please try again.")
        return False
while True:
    print("\n        Hello!\nHow can I help you today?\nInset your PIN and then choose one of the options bellow:")
    print("1. Check your Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit Main Menu")

    if verify_pin():
        choice = input("Enter your choice (1/2/3/4): ")

        if choice == '1':
            check_balance()
        elif choice == '2':
            deposit()
        elif choice == '3':
           print("Withdrawal Options:")
           print("1. $20")
           print("2. $50")
           print("3. $100")
           print("4. Other Amount")
           sub_choice = input("Enter your sub-choice (1/2/3/4): ")

           if sub_choice == '1':
              withdraw(20)
           elif sub_choice == '2':
              withdraw(50)
           elif sub_choice == '3':
              withdraw(100)
           elif sub_choice == '4':
            other_amount=float(input("Enter the withdrawal amount (multiple of 10):$"))
            if other_amount%10==0 and other_amount <= account_balance:
                 withdraw(other_amount)
            elif other_amount > account_balance:
                 print("Insufficient funds.\nYour balance is $", account_balance)
            else:
                 print("Invalid withdrawal amount. Amount must be a multiple of 10")
                 break
        elif choice == '4':
          print("Thank you for using the ATM. Goodbye!")
          break

        another_operation = input("Do you want to perform another operation (yes/no)? ").lower()
        if another_operation != "yes":
          print("Thank you for using the ATM. Goodbye!")
          break



```
